```python
import tensorflow as tf
tf.executing_eagerly()
import numpy as np
from pandas.io.parsers import read_csv

model=tf.compat.v1.global_variables_initializer()

data = read_csv("./data.csv", sep=',')

data
```

```python
NaN = np.isnan(data)
data[NaN] = 0
```

```python
xy = np.array(data, dtype=np.float32)
x_data = xy[:, 1:-1]
y_data = xy[:, [-1]]
```

```python
tf.compat.v1.disable_eager_execution()
```

```python
X = tf.compat.v1.placeholder(dtype = tf.float32, shape=[None, 4])
Y = tf.compat.v1.placeholder(dtype = tf.float32, shape=[None, 1])
W = tf.compat.v1.Variable(tf.compat.v1.random_normal([4,1]), name="weight") 
b = tf.compat.v1.Variable(tf.compat.v1.random_normal([1]), name="bias")
```

```python
hypothesis = tf.matmul(X,W)+b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.000005)
train = optimizer.minimize(cost)
```

```python
sess = tf.compat.v1.Session()
sess.run(tf.compat.v1.global_variables_initializer())
```

```python
for step in range(100001):
    cost_, hypo_, _ = sess.run([cost, hypothesis, train], feed_dict={X:x_data, Y:y_data})
    if step % 500 == 0:
        print("#", step, "Cost of Loss: ", cost_)
        print("Green onions' price :", hypo_[0])
```

```python
saver = tf.compat.v1.train.Saver()
save_path = saver.save(sess, "./saved.cpkt")
print("Finish")
```
